```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('student-mat.csv', sep=';')

print("Shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nData types:")
print(df.dtypes)
```

```python
print("Missing values:")
print(df.isnull().sum())

df = df.drop_duplicates()
print(f"\nShape after removing duplicates: {df.shape}")

print("\nSummary Statistics:")
print(df.describe())
```

```python
print("📊 Average Final Grade (G3):", round(df['G3'].mean(), 2))

above_15 = df[df['G3'] > 15]
print(f"🎯 Students who scored above 15: {len(above_15)}")

correlation = df['studytime'].corr(df['G3'])
print(f"📚 Correlation between study time and grades: {round(correlation, 2)}")

gender_performance = df.groupby('sex')['G3'].mean()
print("\n👥 Average grade by gender:")
print(gender_performance)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Final Grades')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['studytime'], df['G3'], color='coral', alpha=0.6)
plt.title('Study Time vs Final Grade')
plt.xlabel('Study Time')
plt.ylabel('Final Grade (G3)')
plt.show()
```

```python
plt.figure(figsize=(6,5))
gender_performance.plot(kind='bar', color=['pink', 'skyblue'], edgecolor='black')
plt.title('Average Final Grade by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Grade')
plt.xticks(rotation=0)
plt.show()
```
